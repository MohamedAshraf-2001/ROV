```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 

# Define our imshow function 
def imshow(title = "Image", image = None, size = 3):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

Create 4 images, each image has a dimension of 50x50 and a distinct color. After creating the images attach them together to make an image of 100x100, then resize the image to 200x200.

```python
#create 4 black images and merge channels:

image_1 = np.zeros([50,50,3], np.uint8)
B, G, R = cv2.split(image_1)
image_1_merged = cv2.merge([B+255, G, R])          #Blue Image

image_2 = np.zeros([50,50,3], np.uint8)
B, G, R = cv2.split(image_1)
image_2_merged = cv2.merge([B, G+255, R])          #Green Image

image_3 = np.zeros([50,50,3], np.uint8)
B, G, R = cv2.split(image_1)
image_3_merged = cv2.merge([B, G, R+255])          #Red Image

image_4 = np.zeros([50,50,3], np.uint8)
B, G, R = cv2.split(image_1)
image_4_merged = cv2.merge([B+255, G+255, R+255])  #white Image
```

```python
imshow('image_1', image_1_merged)
imshow('image_2', image_2_merged)
imshow('image_3', image_3_merged)
imshow('image_4', image_4_merged)
```

```python
#join images together 
join_1 = cv2.vconcat([image_1_merged, image_2_merged])
join_2 = cv2.vconcat([image_3_merged, image_4_merged])
join_All = cv2.hconcat([join_1,join_2])
imshow('Joint Images', join_All)
join_All.shape
```

```python
#resize it
image_resize = cv2.resize(join_All, (200,200))
imshow('resized image', image_resize)
```
